```
#6
%%manim svo --base64 --low_quality
class svo(Scene):
  def construct(self):
    title=TextMobject(r"\begin{flushleft} {\small INITIALIZATION} \end{flushleft}")
    t0=Text("Let's consider the optimization of the function")
    t0e=TextMobject("$$ -0.5x^2+3x+5 $$")
    title.to_corner(UL)
    t0.to_edge(UP,buff=2)
    t0e.to_edge(UP,buff=3)
    t0e.to_edge(LEFT,buff=5)
    l=Line([0,2.88,0],[1.55,2.88,0],color=RED)
    l2=Line([1.55,2.88,0],[1.55,0,0],color=RED)

    self.play(FadeInFromLarge(title))
    self.play(FadeIn(t0))
    self.play(Write(t0e))
    self.wait(1)
    self.play(FadeOut(t0),FadeOut(title))
    t0e.move_to([-5,3,0])
    self.wait(1)
    self.graph()
    self.play(FadeInFromDown(l))
    self.play(FadeInFromDown(l2))
        
  def graph(self):
        axes = Axes(
            x_min=-10, x_max=10,
            y_min=-12, y_max=12,

            number_line_config={
                "tick_size": 0.05,
            },

            x_axis_config={
                "unit_size": 0.5,
                "tick_frequency": 2,
                "include_numbers": True,
                "numbers_to_show": np.append(np.arange(-10,0,2),np.arange(2,10,2)),
            },
            y_axis_config={
                "unit_size": 0.3,
                "tick_frequency": 2,
                "include_numbers": True,
                "numbers_to_show": [-10,-8,-6,-4,-2,2,4,6,8,10],
                "label_direction": UP,
            },
        )

        self.add(axes)
        self.wait()

        graph = axes.get_graph(
            lambda x: -0.5*pow(x,2)+3*x+5,
            color=YELLOW,
        )

        self.add(graph)
        self.wait()  
```
